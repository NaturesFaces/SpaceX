```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance (lat1, lon1, lat2, lon2):
    #Radius of earth
    R = 6373.0
    lat1 = radians(lat1)
    lat2 = radians(lat2)
    lon1 = radians(lon1)
    lon2 = radians(lon2)

    dlon = lon1 - lon2
    dlat = lat1 - lat2

    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2*atan2(sqrt(a), sqrt(1-a))
    distance = R*c
    return distance
```

```python
distance_coastline = calculate_distance(launch_sites2['Latitude'][1], launch_sites2['Longitude'][1], 28.56288, -80.568)
```

```python
distance_marker = folium.Marker(
    location = (28.56288, -80.568), icon = DivIcon(icon_size = (20, 20), icon_anchor = (0, 0), html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline))
)

lines = folium.PolyLine(locations = ([launch_sites2['Latitude'][1], launch_sites2['Longitude'][1]], [28.56288, -80.568]), weight = 1)
```
